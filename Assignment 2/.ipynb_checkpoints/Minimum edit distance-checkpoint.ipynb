```python
def minimum(x,y,z):
    if(x<y):
        if(x<z):
            return x
        else:
            return z
    else:
        if(y<z):
            return y
        else:
            return z
```

```python
def subs(x,y):
    if(x==y):
        return 0
    return 2
```

```python
# source=input()
# target=intput()
source="intention"
target="execution"
```

```python
n=len(source)
m=len(target)
n=n+1
m=m+1
```

```python
dp=[]
for i in range(0,n):
    dp.append([])
        
```

```python
dp[0].append(0)
```

```python
for i in range(1,n):
    dp[i].append(dp[i-1][0]+1)
for j in range(1,m):
    dp[0].append(dp[0][j-1]+1)
```

```python
for i in range(1,n):
    for j in range(1,m):
        dp[i].append(minimum(dp[i-1][j]+1,dp[i-1][j-1]+subs(source[i-1],target[j-1]),dp[i][j-1]+1))
```

```python
count=0
def backtrack(dp,source,target,i=None,j=None,path=None):
    global count
    if(path==None):
        path=""
    if(i==0 and j==0):
        count=count+1
        print(path)
        print("_________",count,"__________")
        return
    if (i==None):
        i=len(source)
    if (j==None):
        j=len(target)
    if(source[i-1]==target[j-1]):
        backtrack(dp,source,target,i-1,j-1,path)

    if(dp[i-1][j]+1==dp[i][j] and i-1>=0):
        #print("Delete "+source[i-1])
        backtrack(dp,source,target,i-1,j,path+str("Delete "+source[i-1]+"\n"))
    if(dp[i][j-1]+1==dp[i][j] and j-1>=0):
        #print("Insert "+target[j-1])
        backtrack(dp,source,target,i,j-1,path+("Insert "+target[j-1]+"\n"))
    if (dp[i-1][j-1]+2==dp[i][j] and i-1>=0 and j-1>=0):
        #print("Replace ",source[i-1]," with ",target[i-1])
        backtrack(dp,source,target,i-1,j-1,path+("Replace "+source[i-1]+" with "+target[i-1]+"\n"))
```
